```python
import pandas as pd
from datetime import date
bitcoin =pd.read_csv('bitcoin.csv').drop('Unnamed: 0',axis=1)
bitcoin['Date'] = pd.to_datetime(bitcoin['Date'])
#bitcoin.set_index('Date',inplace=True)
ethereum = pd.read_csv('ethereum.csv').drop('Unnamed: 0',axis=1)
ethereum['Date'] = pd.to_datetime(ethereum['Date'])
#ethereum.set_index('Date',inplace=True)
ethereum.head()
```

```python
bitcoin.head()
```

```python
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.models.widgets import Tabs, Panel
from bokeh.io import output_notebook, show,output_file
from bokeh.plotting import figure, show

hover = HoverTool(tooltips={'Date':'@Date{%F}','Price':'@Price'},formatters={'Date':'datetime'})
eth = figure(x_axis_type="datetime",tools=[hover], plot_width=800, plot_height=350,title='Bitcoin Price in USD')
eth.line('Date', 'Price', source=bitcoin,color='Red')
eth.grid.grid_line_alpha = 0.75
eth.xaxis.axis_label = 'Date'
eth.yaxis.axis_label = 'Price'
eth.ygrid.band_fill_color = "olive"
eth.ygrid.band_fill_alpha = 0.1
first = Panel(child=eth, title='bitcoin')


bit = figure(x_axis_type="datetime",tools=[hover], plot_width=800, plot_height=350, title='Ethereum Price in USD')
bit.line('Date','Price',source=ethereum,color='Blue')
bit.grid.grid_line_alpha = 0.75
bit.xaxis.axis_label = 'Date'
bit.yaxis.axis_label = 'Price'
bit.ygrid.band_fill_color = "olive"
bit.ygrid.band_fill_alpha = 0.1

second = Panel(child=bit,title='ethereum')
tabs=Tabs(tabs=[first,second])
output_file('bitcoin&ethereum_fullplots.html')
output_notebook()
show(tabs)
```
