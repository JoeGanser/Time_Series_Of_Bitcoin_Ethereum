```python
import pandas as pd
import numpy as np
from datetime import date
bit = pd.read_csv('bitcoin_data_to_plot.csv').drop('Unnamed: 0',axis=1)
bit.rename(columns={'Predicted with ARIMA':'ARIMA','Predicted with Neural Nets':'Neural_Nets'},inplace=True)
bit_p = bit.loc[7:]
bit['Date'] = pd.to_datetime(bit['Date'])
#bit.set_index('Date',inplace=True)
eth = pd.read_csv('ethereum_data_to_plot.csv').drop('Unnamed: 0',axis=1)
eth.rename(columns={'Predicted with ARIMA':'ARIMA','Predicted with Neural Nets':'Neural_Nets'},inplace=True)
eth_p = eth.loc[7:]
eth['Date'] = pd.to_datetime(eth['Date'])
#eth.set_index('Date',inplace=True)
```

```python
from bokeh.models import ColumnDataSource, HoverTool,Legend
from bokeh.layouts import column, row,widgetbox
from bokeh.models.widgets import Tabs, Panel,DataTable, DateFormatter, TableColumn
from bokeh.io import output_notebook, show,output_file
from bokeh.plotting import figure, show
bit.tail()
```

```python
source1 = ColumnDataSource(bit)
source1_p = ColumnDataSource(bit_p.loc[7:])
hover = HoverTool(tooltips={'Date':'@Date{%F}','Price':'@Price','ARIMA':'@ARIMA','Neural Nets':'@Neural_Nets'},formatters={'Date':'datetime'})
bitcoin = figure(x_axis_type="datetime",tools=[hover], plot_width=700, plot_height=350, title='Bitcoin Price Predictions August 2017')
bitcoin.line('Date',list(bit.columns)[1],source=source1,color='Blue',legend='Actual')
bitcoin.square('Date',list(bit.columns)[2],source=source1,color='Red',legend ='ARIMA prediction')
bitcoin.circle('Date',list(bit.columns)[3],source=source1,color='Black',legend='Neural Nets prediction')
bitcoin.grid.grid_line_alpha = 0.75
bitcoin.xaxis.axis_label = 'Date'
bitcoin.yaxis.axis_label = 'Price'
bitcoin.ygrid.band_fill_color = "grey"
bitcoin.ygrid.band_fill_alpha = 0.1
bitcoin.legend.location='top_left'

columns = [TableColumn(field='Date', title="Date"),
           TableColumn(field="Price", title="Price"),
           TableColumn(field="ARIMA", title="ARIMA Predictions"),
           TableColumn(field="Neural_Nets", title="Neural Nets Predictions")
          ]
data_table1 = DataTable(source=source1_p, columns=columns, width=700, height=300)
bitcoin_plots = column([bitcoin,data_table1])
output_notebook()
show(bitcoin_plots)
```

```python
source2 = ColumnDataSource(eth)
source2_p = ColumnDataSource(eth_p)
hover = HoverTool(tooltips={'Date':'@Date{%F}','Price':'@Price','ARIMA':'@ARIMA','Neural Nets':'@Neural_Nets'},formatters={'Date':'datetime'})
ethereum = figure(x_axis_type="datetime",tools=[hover], plot_width=700, plot_height=350, title='Ethereum Price Predictions August 2017')
ethereum.line('Date',list(eth.columns)[1],source=source2,color='Blue',legend='Actual')
ethereum.square('Date',list(eth.columns)[2],source=source2,color='Red',legend ='ARIMA prediction')
ethereum.circle('Date',list(eth.columns)[3],source=source2,color='Black',legend='Neural Nets prediction')
ethereum.grid.grid_line_alpha = 0.75
ethereum.xaxis.axis_label = 'Date'
ethereum.yaxis.axis_label = 'Price'
ethereum.ygrid.band_fill_color = "grey"
ethereum.ygrid.band_fill_alpha = 0.1
ethereum.legend.location='top_left'

columns = [TableColumn(field='Date', title="Date"),
           TableColumn(field="Price", title="Price"),
           TableColumn(field="ARIMA", title="ARIMA Predictions"),
           TableColumn(field="Neural_Nets", title="Neural Nets Predictions")
          ]
data_table2 = DataTable(source=source2_p, columns=columns, width=700, height=300)
ethereum_plots = column([ethereum,data_table2])
output_notebook()
show(ethereum_plots)
```

```python
from bokeh.models.widgets import Tabs, Panel
b = Panel(child=bitcoin_plots, title='Bitcoin Predictions')
e = Panel(child=ethereum_plots, title='Ethereum Predictions')
t=Tabs(tabs=[b,e])
output_file('bitcoin_ethereum_prediction_plots.html')
show(t)
```
